```python
# mementukan outliner dari kolom price
q1_price = df.reviews_per_month.quantile(.25)
q3_price = df.reviews_per_month.quantile(.75)
iqr_price = q3_price - q1_price
print('IQR dari kolom price: ', iqr_price)
                                   
# menentukan limit atas dan bawah
upl_reviews_per_month = q3_price + 1.5 * iqr_price
lwl_reviews_per_month = q1_price - 1.5 * iqr_price

print('Limit atas kolom price: ', upl_reviews_per_month)
print('Limit bawah kolom price: ', lwl_reviews_per_month)
```

* kolom price dan minimum_nights memiliki outliner
* price > 75% dan max memiliki jarak yang jauh serta nilai min = 0 (tidak mungkin ada rental yang price nya 0)
* minimum_nights > 75% dan max memiliki jarak yang jauh. standar properti dapat dikatakan penginapan juka pemilik menyewakan properti dalam waktu kurang dari 30 hari. maka minimum_nights >30 dianggap bukan penginapan 

```python
# mementukan outliner dari kolom price
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
iqr_price = q3_price - q1_price
print('IQR dari kolom price: ', iqr_price)
                                   
# menentukan limit atas dan bawah
upl_price = q3_price + 1.5 * iqr_price
lwl_price = q1_price - 1.5 * iqr_price

print('Limit atas kolom price: ', upl_price)
print('Limit bawah kolom price: ', lwl_price)
```

***cleansing after EDA***


```python
# membuat duplikay datafame
df_new = df.copy(deep=True)
```

```python
print(len(df))
```

```python
#hapus data yang tidak perlu hasil dari EDA

df_new.drop(df_new[df_new.price > upl_price].index, axis=0, inplace=True)
df_new.drop(df_new[df_new.price == 0.0].index, axis=0, inplace=True)
df_new.drop(df_new[df_new.minimum_nights > 30].index, axis=0, inplace=True)

print('Jumlah data sebelum dibersihkan: ', len(df))
print('Jumlah data setelah dibersihkan: ',len(df_new))
print(f'{round(((1 - (df_new.shape[0]/df.shape[0]))*100), 2)}% data telah dihapus.')
```

```python
#replace data null pada kolom last_review dan reviews_per_month

df_new['last_review'].fillna(0, inplace=True)
df_new['reviews_per_month'].fillna(0, inplace=True)
```

```python
room_type_counts = df_new['room_type'].value_counts()

plt.bar(room_type_counts.index, room_type_counts.values, color='skyblue')

for i, value in enumerate(room_type_counts.values):
    plt.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.xlabel('Room Type')

plt.ylabel('Count')
plt.title('Bar Chart of Room Type')
plt.show()
```

```python
plt.figure(figsize=(12,6));
sns.set_theme()
custom_palette = {'Entire home/apt': 'red', 'Private room': 'yellow', 'Shared room': 'blue', 'Hotel room': 'green'}

ax = sns.barplot(data=df_new,
            x='room_type',
            y='number_of_reviews',
            estimator=np.sum,
            ci=0,
            palette=custom_palette);
for bar in ax.patches:
    ax.annotate(f'{bar.get_height():,.0f}',
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                ha='center', va='center', size=10, xytext=(0, 8),
                textcoords='offset points')

plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Average price per room type');
```

```python
df_new['neighbourhood'].value_counts()
```

```python
total_reviews_per_neighborhood = df_new.groupby('neighbourhood')['number_of_reviews'].sum().reset_index()
df_sorted = total_reviews_per_neighborhood.sort_values(by='number_of_reviews', ascending=False)
top_5_neighborhoods = df_sorted.head(5)

plt.figure(figsize=(12, 6))
sns.set_theme()

ax = sns.barplot(data=top_5_neighborhoods,
                 x='neighbourhood',
                 y='number_of_reviews',
                 estimator=np.sum,
                 ci=0,
                 color='red')

for bar in ax.patches:
    ax.annotate(f'{bar.get_height():,.0f}',
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                ha='center', va='center', size=10, xytext=(0, 8),
                textcoords='offset points')

plt.xlabel('Neighbourhood')
plt.ylabel('Number of Reviews')
plt.title('Top 5 Neighbourhoods with the Most Popular')

plt.show()
```
