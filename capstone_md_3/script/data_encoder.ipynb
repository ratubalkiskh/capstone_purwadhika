```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv(r'D:\Purwadhika\belajar aja\learning\capstone_3\data\data_travel_insurance_prep.csv',sep='|')
df.head()
```

```python
category_columns = df[['Agency', 'Agency Type', 'Distribution Channel', 'Product Name', 'Claim', 'Destination']] 
#mengambil kolom2 kategorical
```

```python
label_encoder = LabelEncoder() #memanggil model label encoder

category_transform = category_columns.apply(LabelEncoder().fit_transform) 
# melakukan transformasi kolom category kedalam bentuk angka

category_transform
```

```python
df.drop(['Agency', 'Agency Type', 'Distribution Channel', 'Product Name', 'Claim', 'Destination'], axis = 1, inplace = True) 
#hapus kolom kategori pada df utama
df = pd.concat([df, category_transform], axis=1) #menyatukan df utama dengan df hasil transform label
df
```

```python
df.to_csv(r'D:\Purwadhika\belajar aja\learning\capstone_3\data\data_travel_insurance_transform.csv',sep='|', index=False)
```
